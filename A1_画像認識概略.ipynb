# 1 画像認識とは

- 画像認識(Image Recognintion）とは、画像や動画に写る場所や人物など、様々なオブジェクトを認識する技術を指します。
- 近年ニュースを騒がしている自動運転なども画像認識技術を活用しています。
- また、レントゲン写真を解析して特定の症状/病気を検知するなど、幅広い分野で大きな期待が寄せられています。
- 人間の目と脳のように、コンピュータに画像を認識させることが画像認識です。

## 1-1 コンピュータビジョンとは

- 画像認識は「コンピュータビジョン」の一部の分野です。
- コンピュータビジョン（Computer vision）とは、コンピュータに画像や動画を入力して、必要な情報（顔認識や物体検知）を取り出す技術です。
- つまり、コンピュータが画像をいかに理解できるかを扱う技術です。
- コンピュータービジョンは画像認識に加えて、イベント検知や動画内の特定オブジェクトを追跡するビデオトラッキングなども含まれます。


## 1-2 画像認識技術の歴史

- 画像認識技術の歴史は長く、1960年代から研究が行われています。
- 研究が開始した第一次AIブーム（1960年代）は、コンピュータの処理能力に限界があり極めて限定的な技術でした。
- その後、コンピュータの処理能力は劇的に改善され、画像認識技術の応用範囲も広まりました。
- 近年は第三次AIブームと位置付けられていますが、そのきっかけとなったのも画像認識がきっかけと言われています。

## 1-3 ILSVRC

- 画像認識が世間一般に大きく注目されたのが2012年に開催された「ILSVRC」です。
- ILSVRCとはImageNet Large Scale Visual Recognition Challengeの略です。
- 2010年から始まった大規模データを用いた画像認識の競技会です。
- それまでは最先端の上位画像認識技術でも70%前半の正解率でした。
- 2012年の大会でディープラーニングを利用した画像認識が登場し、約85%の正解率を叩き出し世間を大きく騒がしました。その年は1位と2位の差が10%以上も離れた特別な年でした。
- ディープラーニングの研究およびコンピュータの処理能力の改善により、2015年の大会では人間の画像認識能力をコンピュータが超えました。少し耳を疑うような話ですが、これは人間が画像認識するよりもコンピューターが画像認識をしたほうが精度が高いことを示します。
- 画像認識の適用範囲は非常に多岐に渡ります。Facebookの顔認識アルゴリズムは98%の正解率で顔を識別することが可能です。少し前までは未来の象徴として見られていた「自動運転」も、今では世界中で実証実験が行われています。

# 2 画像認識の仕組み

画像認識、つまりコンピュータが「画像を認識」するとありますが、どのように認識するのでしょうか？コンピュータは私たち人間のように画像を画像のまま見ません。

コンピュータは画像をピクセルの値で扱います。ピクセルとは画像がもつ色情報の最小の要素です。一般的なカラー画像においては1つのピクセルに三原色（赤、緑、青）の独立した「輝度」の情報を保有します。図2.1の画像は白と黒のみのグレースケールですので、1層のピクセルに輝度情報が与えられています。

ピクセルにどれだけの情報量を与えるかによってピクセルが表現できる色数が決まります。白黒の画像は8ビットですので0から最大255までの値で輝度を表します。

画像の白い部分のピクセルは数値が低いのがわかります。対して画像の黒い部分は数値が高いです。コンピュータが画像を認識するとありますが、決して人間のように画像を見て「理解」している訳ではありません。コンピューターは画像を構成するピクセルの値を「解析して特定のパターンなどを学習」するのです。

# 3 画像認識の活用事例

## 3-1 自動画像整理機能

画像認識技術が最も活躍している機能の一つが画像の整理整頓です。スマートフォンの普及により、私たちの生活の中でデジタル写真を撮影する機会が劇的に増えています。

クラウド上で写真を管理出来るGoogleフォトには画像認識を利用した驚くべき機能が搭載されています。その一つの例が「Enhanced Sharing」です。Googleフォトへアップロードされた写真は画像認識技術により、その写真に写っていると思われる人物を特定して共有を促してくれます。

## 3-2 画像検索

検索といえば従来は「キーワード」による検索が一般的でした。画像認識技術が発達したことにより、近年では画像を使った検索も広く使われるようになりました。

具体的な事例として「カーセンサー 中古車検索アプリ」を紹介します。

カーセンサーでは約８０万枚の車の画像を事前に用意をして、ディープラーニングを応用した画像認識を行いました。結果として車の画像から車種判別が行える検索機能をアプリに実装しています。従来の検索では車の車種が解らないと検索が行えませんでした。画像認識技術により、車種名が解らなくても画像から検索することが可能になります。

## 3-3 画像認識による医療診断

画像認識による医療診断
GoogleのCEOピチャイ氏は「（医療は）AI（機械学習）が革新をもたらす最も重要な領域」と発言して話題になりました。画像認識は医療でも多岐の分野で広く活用されています。

その一つがGoogleが発表した「糖尿病性網膜症」の前兆検知です。あまり馴染みのある病名ではありませんが、日本人成人の失明で最も高い原因となる病気です。Googleは網膜画像を画像認識技術で解析を行い、非常に高い精度で失明の前兆を検出することに成功しました。

## 3-4 自動車安全装備

自動車が発明されてから今日に到るまで様々な安全装置が発明されました。シートベルトやエアーバッグは従来から広く普及している安全装備ですが、これらに加えて近年では画像認識を活用した安全装備が登場しています。

東芝が開発・販売をしている画像認識プロセッサ「Viscontiシリーズ」は、まさに画像認識を活用した自動車安全装備の一つです。このプロセッサは自動車に搭載されたカメラ画像やレーダー反射波などから障害物を検知します。障害物を検知することで、危険と判断したらブレーキをかけ交通事故を未然に防いでくれます。
